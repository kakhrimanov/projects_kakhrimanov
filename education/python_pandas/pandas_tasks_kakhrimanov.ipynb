Частный вкладчик разместил в банке пять депозитов на разные сроки. 
В серии amounts представлены суммы депозитов в рублях. 
В серии days перечислены сроки размещения в днях. 
Серия rates содержит ставки в процентах.

Посчитайте суммы процентов, которые получит вкладчик по каждому депозиту. 
База для расчета процентов = 365 дней. 
Округлите результат до двух знаков после запятой 
и выведите итоговый объект Series на печать.

```python
import pandas as pd

amounts = [300_000, 200_000, 500_000, 700_000, 1_000_000]
days = [30, 60, 90, 120, 180]
rates = [10, 11, 11, 12, 14]
amounts = pd.Series(amounts, name='Сумма депозита')
days = pd.Series(days, name='Дни')
rates = pd.Series(rates, name='Ставки')
# Продолжите написание кода здесь
revenue = amounts.mul(rates).div(100).mul(days).div(365).round(2)
print(revenue)
```
