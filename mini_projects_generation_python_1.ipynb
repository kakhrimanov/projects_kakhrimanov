Пользователь потенциально может ввести неверные данные, например, не число, или число превышающее 
100. Важно предусмотреть такую возможность, чтобы программа продолжала правильно работать. 
Обработка такого рода ситуаций называется защитой от дурака.

Напишите функцию is_valid() в которую передается один строковый аргумент. 
Функция возвращает значение True если переданный аргумент является целым числом от 
1 до 100 и False в противном случае. 

1) Организуйте цикл, который будет запрашивать у пользователя данные (цикл может быть бесконечным (while True) 
    или может использовать сигнальную метку с последующим переключением, после угадывания числа);
2) Запросите у пользователя число от 1 до 100;
3) Проверьте введенные данные с помощью функции is_valid():
если данные некорректны, выведите текст: 'А может быть все-таки введем целое число от 1 до 100?' 
    и перейдите к следующей итерации основного цикла;
если данные корректны, преобразуйте их к целому числу для удобства дальнейшей работы.

1) Организуйте сравнение введенного числа с загаданным числом:
    Если введенное число меньше загаданного числа, выведите текст: 'Ваше число меньше загаданного, попробуйте еще разок';
    Если введенное число больше загаданного числа, выведите текст: 'Ваше число больше загаданного, попробуйте еще разок';
    Если введенное число равно загаданному числу, выведите текст: 'Вы угадали, поздравляем!'.
2) Выведите прощальное сообщение пользователю: 'Спасибо, что играли в числовую угадайку. Еще увидимся...'.

Добавьте подсчет попыток, сделанных пользователем. Когда число отгадано, программа должна показать количество попыток;
Добавьте возможность генерации нового числа (повторная игра), после того, как пользователь угадал число;
Добавить возможность указания правой границы для случайного выбора числа (от 1 до n).

```python
import random 

# Угадайка
def guess_num():
    print('Введите правую границу для случайного выбора числа')
    n = int(input()) # Указание правой границы для случайного выбора числа

    # Функция, проверяющая поданный текст к нужному формату (число в диапазоне от 1 до n)
    def is_valid(text):
        return (text.isdigit() and 1 <= int(text) <= n)

    # Функция, повторяющая ввод числа, до тех пор, пока не получим число от 1 до n
    def input_num():
        # Цикл продолжается до тех пор, пока пользователь не введет число от 1 до n
        while True:
            num = input() # Ввод числа от пользователя
            if is_valid(num):
                num = int(num) # Преобразуем в целочисленный тип
                return num
            else:
                print('А может быть все-таки введем целое число от 1 до {}?'.format(n)) 

    random_num = random.randint(1, n) # Загаданное число
    print('Добро пожаловать в числовую угадайку')


    print('Введите число от 1 до {}'.format(n))
    count = 0 # Подсчет числа попыток
    num = input_num() # Принимаемое число к отгадке
    count = 1  # Случилась первая попытка

    # Пока, не отгадаем число, будем пытаться и дальше отгадывать
    while num != random_num:
        if num < random_num:
            print('Ваше число меньше загаданного, попробуйте еще разок')
        else:
            print('Ваше число больше загаданного, попробуйте еще разок')
        num = input_num()
        count += 1
    
    print(f'Вы угадали, поздравляем! Число попыток = {count}')
    print('Спасибо, что играли в числовую угадайку. Еще увидимся...')


guess_num() # Вызов угадайки в первый раз


# Цикл на повторение игры
while True:
    print('Хотите сыграть ещё? Напишите "Продолжить"')
    answer = input()
    
    if answer == 'Продолжить':
        guess_num()
    else:
        print('Хорошего дня!')
        break
```

# Генератор безопасных паролей (№3)

Подключите модуль random;
Создайте строковые константы:
    digits: 0123456789;
    lowercase_letters: abcdefghijklmnopqrstuvwxyz;
    uppercase_letters: ABCDEFGHIJKLMNOPQRSTUVWXYZ;
    punctuation: !#$%&*+-=?@^_.
Создайте переменную chars = '', которая будет содержать все символы, которые могут быть в генерируемом пароле.

Программа должна запрашивать у пользователя следующую информацию:
    Количество паролей для генерации;
    Длину одного пароля;
    Включать ли цифры 0123456789?
    Включать ли прописные буквы ABCDEFGHIJKLMNOPQRSTUVWXYZ?
    Включать ли строчные буквы abcdefghijklmnopqrstuvwxyz?
    Включать ли символы !#$%&*+-=?@^_?
    Исключать ли неоднозначные символы il1Lo0O?
    
На основании введенной пользователем информации, сформируйте переменную chars, содержащую все символы, которые могут быть в генерируемом пароле.

Напишите функцию generate_password(), которая принимает два аргумента:
    length: длину пароля;
    chars: алфавит из символов которого состоит пароль;
    и возвращает пароль.
Используя цикл for, сгенерируйте необходимое количество паролей.

```python
import random

# Строковые константы
digits = '0123456789'
uppercase_letters = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
lowercase_letters = 'abcdefghijklmnopqrstuvwxyz'
punctuation = '!#$%&*+-=?@^_.'

chars = ''

# Возвращает натуральное число
def is_correct(text_error):
    # Запускает цикл, до тех пор, пока не будет введено натуральное число
    while True:
        var = input()
        if var.isdigit() and int(var) > 0:
            var = int(var)
            return var
        else:
            print(text_error)
    
# Возвращает Булево в зависимости от ответа (да/нет)
def is_correct_response_cycle(ans):
    # Запускает цикл, до тех пор, пока не будет введено да или нет
    while ans != 'да' and ans != 'нет':
        print('Напишите да или нет')
        ans = input()
    return (ans == 'да')

# Возвращает случайный пароль в виде массива символов
def generate_password(length, chars):
    return random.sample(chars, length)
    
    
print('Введите количество паролей для генерации')
n = is_correct(text_error='Введите положительное целочисленное количество паролей для генерации')
    
print('Введите длину одного пароля') 
length = is_correct(text_error='Введите положительную целочисленную длину пароля')


# Массив вопросов
arr_questions = [
    'Включать ли цифры 0123456789? Напишите да или нет', 
    'Включать ли прописные буквы ABCDEFGHIJKLMNOPQRSTUVWXYZ? Напишите да или нет',
    'Включать ли строчные буквы abcdefghijklmnopqrstuvwxyz? Напишите да или нет',
    'Включать ли символы !#$%&*+-=?@^_? Напишите да или нет',
    'Исключать ли неоднозначные символы il1Lo0O?  Напишите да или нет'                               
]

arr_flags = list() # Массив полученных ответов (True или False, то есть, да или нет)
for i in range(len(arr_questions)):
    print(arr_questions[i])
    arr_flags.append(is_correct_response_cycle(ans=input()))
    
arr_str_cons = [digits, uppercase_letters, lowercase_letters, punctuation] # Массив строковых констант

for i in range(len(arr_str_cons)):
    if arr_flags[i]:
        chars += arr_str_cons[i] # Если ответ был True, то добавляем строковую константу в chars
    
    
if arr_flags[-1]: # Если на последний вопрос был дан ответ да (True), то исключаем данные символы из chars
    for char in chars: 
        if char in 'il1Lo0O':
            chars = chars.replace(char, '')

# print(chars) # для отладки. 
    
# Генерируем необходимое количество паролей
for i in range(n):
    print(*generate_password(length, chars), sep='')
```

# Шифр Цезаря (№4)

На вход программе подаётся строка текста на английском языке, 
в которой нужно зашифровать все слова. Каждое слово строки следует 
зашифровать с помощью шифра Цезаря (циклического сдвига на длину этого слова). 
Строчные буквы при этом остаются строчными, а прописные – прописными. 
Гарантируется, что между различными словами присутствует один пробел.

```python
def cipher_char(char, n): # Шифруем букву, подавая саму букву и сдвиг
    
    s2 = '' # сюда будет вставлена зашифрованная буква
    en = 'abcdefghijklmnopqrstuvwxyz' # строчный английский алфавит
    EN = 'abcdefghijklmnopqrstuvwxyz'.upper() 
    
    # Коды букв
    a = ord('a') 
    z = ord('z')
    A = ord('A')
    Z = ord('Z')
    
    # Относится ли символ к строчному английскому алфавиту
    if char in en: 
        order = ord(char) + n # Код нового символа 
        if order > z: # Если код нового символа выходит за границы строчного алфавита
            s2 += chr((order - z) + a - 1) # то начинаем с самого начала строчного алфавита
        else:
            s2 += chr(order)
    # По аналогии со строчным алфавитом
    elif char in EN:
        order = ord(char) + n
        if order > Z:
            s2 += chr((order - Z) + A - 1)
        else:
            s2 += chr(order)
    # Если символ, не буква 
    else:
        s2 += char
    return s2

# Подсчитывает количество букв в слове 
def count_alpha(word):
    count = 0
    for char in word:
        if char.isalpha():
            count += 1
    return count


words = input().split() # Подаем строку текста, разбивая на пробелы. 

new_words = list() # Массив зашифрованных слов
for word in words: # Для каждого слова
    length = count_alpha(word) # Получаем количество БУКВ
    string = ''
    for char in word: # Для каждого символа в слове
        if char.isalpha(): # Если символ - буква
            char_ch = cipher_char(char, length) # Получаем зашифрованную букву
            string += char_ch # и добавляем её в новое слово
        else:
            string += char # добавляем символ (не букву) в новое слово
    new_words.append(string) # Добавляем новое слово
    
print(*new_words)
```

# Угадайка слов (№5)

Подключите модуль random;
Создайте глобальный список word_list, содержащий слова, которые будут использоваться в игре
Напишите функцию get_word() которая возвращает случайное слово из списка word_list в верхнем регистре.
Написать минипроект по угадыванию слов.

```python
# функция получения текущего состояния
def display_hangman(tries):
    stages = [  # финальное состояние: голова, торс, обе руки, обе ноги
                '''
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / \\
                   -
                ''',
                # голова, торс, обе руки, одна нога
                '''
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / 
                   -
                ''',
                # голова, торс, обе руки
                '''
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |      
                   -
                ''',
                # голова, торс и одна рука
                '''
                   --------
                   |      |
                   |      O
                   |     \\|
                   |      |
                   |     
                   -
                ''',
                # голова и торс
                '''
                   --------
                   |      |
                   |      O
                   |      |
                   |      |
                   |     
                   -
                ''',
                # голова
                '''
                   --------
                   |      |
                   |      O
                   |    
                   |      
                   |     
                   -
                ''',
                # начальное состояние
                '''
                   --------
                   |      |
                   |      
                   |    
                   |      
                   |     
                   -
                '''
    ]
    return stages[tries]
```

```python
def play(word):
    # Является ли пользователь дураком
    def is_fool(text):
        # Если текст, не является буквой/словом или данная буква или данное слово было уже названо
        if not text.isalpha() or text in guessed_letters or text in guessed_words: 
            return True
        else:
            return False
    
    print('Давайте играть в угадайку слов!')
    word_completion = '_' * len(word)  # строка, содержащая символы _ на каждую букву задуманного слова   
    
    guessed_letters = []               # список уже названных букв
    guessed_words = []                 # список уже названных слов
    guessed = False                    # сигнальная метка
    
    tries = 6 # Количество попыток
    

    while tries > 0:
        print(display_hangman(tries)) # Состояние игры
        print(word_completion)
        print('Введите пожалуйста букву или слово')
        text = input().upper()

        # Пока пользователь не введет корректно текст
        while is_fool(text):
            print('Введите пожалуйста букву или слово, которое не было названо')
            text = input().upper()



        if text == word:
            guessed = True
            break # Выходим из цикла
        elif text in word:
            guessed_letters.append(text) # Добавляем символ в массив названных букв
            ind = [i for i in range(len(word)) if word[i] == text] # Находим индексы этого символа в загадочном слове

            word_completion = list(word_completion) # Преобразуем в список зашифрованное слово
            # Заменяя в нем _ на нужный символ
            for i in range(len(word_completion)):
                if i in ind:
                    word_completion[i] = text 
            word_completion = ''.join(word_completion) # Возвращаем обратно зашифрованное слово в строку
            
            if word_completion.count('_') == 0:
                guessed = True
                break # Выходим из цикла
            else:
                print('Угадали букву')
            
            
        else:
            # Иначе, не угадали и попытка сгорела
            if len(text) == 1:
                guessed_letters.append(text)
                print('Не угадали букву')
            else:
                guessed_words.append(text)
                print('Не угадали слово')
            tries -= 1
    
    return guessed

```

```python
import random 
world_list = ['apple', 'banana', 'pineapple', 'mango', 'plum']

# Получает одно случайное слово из списка
def get_word():
    w = random.choice(world_list).upper()
    return w

random_word = get_word() # Программа загадала слово
if play(random_word):
    print('Поздравляем, вы угадали слово! Вы победили!')
else:
    print(f'Попытки исчерпаны, загадочное слово: {random_word}')
    print(display_hangman(0))
    
```

```python
s1 = 'To be, or not to be, that is the question!'
en = 'abcdefghijklmnopqrstuvwxyz' # 97 и 122
EN = 'abcdefghijklmnopqrstuvwxyz'.upper()
n = 17
s2 = ''

for char in s1:
    if char in en:
        order = ord(char) + n
        if order > 122:
            s2 += chr((order - 122) + 97 - 1)
        else:
            s2 += chr(order)
    elif char in EN:
        order = ord(char) + n
        if order > ord('Z'):
            s2 += chr((order - ord('Z')) + ord('A') - 1)
        else:
            s2 += chr(order)
    else:
        s2 += char
print(s2)
```

Расшифруйте данный текст.

```python
s1 = 'Sgd fqzrr hr zkvzxr fqddmdq nm sgd nsgdq rhcd ne sgd edmbd.'
en = 'abcdefghijklmnopqrstuvwxyz' # 97 и 122
EN = 'abcdefghijklmnopqrstuvwxyz'.upper()
n = 25
s2 = ''
a = ord('a')
z = ord('z')
A = ord('A')
Z = ord('Z')

for char in s1: # Для символа в тексте
    if char in en: # Относится ли символ к строчному английскому алфавиту
        order = ord(char) - n # Порядок символа, вычитая сдвиг 
        if order < a:
            s2 += chr((order + z) - a + 1)
        else:
            s2 += chr(order)
    elif char in EN:
        order = ord(char) - n
        if order < A:
            s2 += chr((order + Z) - A + 1)
        else:
            s2 += chr(order)
    else:
        s2 += char
print(s2)
        
    
```

Приведённый ниже текст был получен в результате шифрования алгоритмом Цезаря со сдвигом вправо на n символов:
Hawnj pk swhg xabkna ukq nqj.

Переведите десятичное число 1000 в шестнадцатеричную систему счисления.

```python
x1 = 1000 // 16
y1 = 1000 % 16
x2 = x1 // 16
y2 = x1 % 16
x3 = x2 // 16
y3 = x2 % 16

print(x2)
print(y3, y2, y1)
```

Переведите десятичное число 513 в двоичную систему счисления.

```python
num10 = 513

n = 2 # Двоичная система

res = ''

while num10 >= n:
    r = num10 % n
    res = str(r) + res
    num10 //= n

print(num10)
res = str(num10) + res
    
print(res)
```
